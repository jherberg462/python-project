```python
#dependencies
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv('data/cleaned_data.csv')
df.head(1)
```

```python
#remove video_error_or_removed column due to previously removing all videos with a True value
df = df.drop('video_error_or_removed', axis=1)
df.head(1)
```

```python
#convert strings on trending_date and publish_time to datetime format that can be calculated in pandas
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
df['publish_time'] = pd.to_datetime(df['publish_time'], format='%Y-%m-%d')
df['publish_time'] = df['publish_time'].dt.tz_localize(None) #remove timezone from publish_time
df['publish_time'] = df['publish_time'].dt.date #convert publish_time into date object
df['trending_date'] = df['trending_date'].dt.date #convert trending_date column into date object
df.head(5)
```

```python
#add column that is the number of days between when the video was trending and the date the video was published
df['Days to trend'] = df['trending_date'] - df['publish_time']
df.head(20)
```

```python
#export to csv file
df.to_csv('data/cleaned_data.csv')
```
